# Test Lagrange Polynomials

```julia
using FastGaussQuadrature
using GeometricIntegrators
using Gadfly
```

Number of nodes:

```julia
s = 3
```

Number of points to plot:

```julia
np = 101
```

Obtain Gauss-Legendre nodes:

```julia
gl = gausslegendre(s)
c  = (gl[1]+1)/2
```

Define Lagrange polynomial:

```julia
lag = LagrangePolynomial(c, ones(s))
```

Compute plot range:

```julia
xp = collect(linspace(0., 1., np))
```

Plot all basis polynomials:

```julia
layers = []

for i in 1:s
    y = zeros(s)
    y[i] = 1
    yp = zeros(np)
    evaluate!(lag, y, xp, yp)
    push!(layers, layer(x=xp, y=yp, Geom.line))
end

plot(layers...)
```
